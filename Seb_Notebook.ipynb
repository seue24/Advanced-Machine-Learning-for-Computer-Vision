- Structural Similarity Index Measure (SSIM)
    - Combines image structure, luminance, and contrast which are calculated locally for each pixel
    - Variants: Multi-scale SSIM which calculates local luminance, contrast and structure additionally for four downscaled versions of the images and combines them in a weighted fashion
                Complex-wavelet SSIM designed to compensate for small rotations and spatial translations
    - Requires ground truth reference image for each output
 - Peak Signal-to-Noise Ratio (PSNR) 
    - developed to measure reconstruction quality of a lossy compressed image compared to the uncompressed reference image
    - Often used as a metric for assessing image similarity
    - requires ground truth reference image for each output so we can#t use it
- Error metrics such as MAE, MSE, RMSE, NMSE 
    - not really an option for us since they directly depend on the absolute difference of intensity values at equal pixel locations
- Learned perceptual image patch similarity (LPIPS)
    - Passes images through a pretrained deep network and comparing their feature representations at multiple layers
    - Relies on feature maps from a trained image classification model
    - Computes difference between feature maps of image1 and image2 usually L2 difference channel-wise
    - Adaption: Deep Image Structure and Texture Similarity (DISTS) giving more focus on texture
- Mutual information (MI) estimates the amount of information of an image R that can be predicted from image I
    - Probably not suitable because we don#t want to retain information
- Frechet Inception Distance (FID)
    - Evaluates the distance between the generated images and real images in the feature space
    - Widely used in GAN papers including unpaired translation works
- Kernel Inception Distance (KID)
    - Distribution divergence metric between generated and real images like FID
    - Computes the maximum mean discrepancy in feature space which is a measure of how different two sets of samples are using a kernel function ondeep features
- Inception Score
    - Inception evaluates quality and diversity of a set of generated images without using any real images
    - One of the earliest GAN metrics
- NIQE
    - Builds a model of Natural Scene Statistics from a corpus of natural images and then computes statistical features
    - When evaluating a test image, NIQE extracts the same features and compares them to the model of natural image features
- BRISQUE Blind/referenceless Image Spatial Quality Evaluator 
    - Also relies on Natural Scene Statistics but unlike NIQE which is completly unsupervised BRISQUE was developed as a distortion-generic but trained model
    - Computes a set of features from locally normalized luminance patches of the image and these feature capture deviations like blurring noise etc by quantifying loss of naturalness in luminance patterns
    - BRISQUE then uses a regression modelto map these features to a quality score
- Precision and Recall for Generative models
- Human evaluation
